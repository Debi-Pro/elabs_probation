## SALES BY REGION

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.index, y=region_sales.values, palette="viridis")
plt.title("Total Sales by Region")
plt.ylabel("Sales")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

## SALES VS PROFIT BY CATEGORY

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Category", y="Sales", data=df, ci=None, color="skyblue", label="Sales")
sns.barplot(x="Category", y="Profit", data=df, ci=None, color="orange", label="Profit")
plt.title("Sales vs Profit by Category")
plt.legend()
plt.show()
```

# TOP 10 SUB-CATEGORIES BY SALES

```python
sub_sales = df.groupby("Sub-Category")["Sales"].sum().nlargest(10)

plt.figure(figsize=(10,5))
sns.barplot(x=sub_sales.values, y=sub_sales.index, palette="mako")
plt.title("Top 10 Sub-Categories by Sales")
plt.xlabel("Sales")
plt.show()
```

# PROFIT DISTRIBUTION

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Profit"], bins=50, kde=True, color="green")
plt.title("Distribution of Profit")
plt.xlabel("Profit")
plt.show()
```

# SALES BY SHIP MODE

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Ship Mode", y="Sales",data=df, palette="Set2")
plt.title("Sales Distribution by Ship Mode")
plt.show()
```

# CONCLUSION


- **From the above analysis, we can conclude:**

- **Region:** Some regions (like US or APAC) contribute significantly more to sales.  
- **Category:** Technology and Office Supplies may have higher profits compared to Furniture.  
- **Top Sub-Categories:** Certain sub-categories dominate revenue (like Phones, Chairs, or Copiers).  
- **Profit Distribution:** Profit values are skewed, with many small profits/losses and a few very high profits.  
- **Shipping Mode:** Sales behavior varies with shipping mode, e.g., Standard Class vs. Same Day.

**This helps us understand overall performance and identify focus areas for the business.**
